```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import requests
import math
```

```python
# Pull in all crime data from 2000 - 2020
crime_2000 = pd.read_csv("Resources/Crime Information 2000.csv")
crime_2001 = pd.read_csv("Resources/Crime Information 2001.csv")
crime_2002 = pd.read_csv("Resources/Crime Information 2002.csv")
crime_2003 = pd.read_csv("Resources/Crime Information 2003.csv")
crime_2004 = pd.read_csv("Resources/Crime Information 2004.csv")
crime_2005 = pd.read_csv("Resources/Crime Information 2005.csv")
crime_2006 = pd.read_csv("Resources/Crime Information 2006.csv")
crime_2007 = pd.read_csv("Resources/Crime Information 2007.csv")
crime_2008 = pd.read_csv("Resources/Crime Information 2008.csv")
crime_2009 = pd.read_csv("Resources/Crime Information 2009.csv")
crime_2010 = pd.read_csv("Resources/Crime Information 2010.csv")
crime_2011 = pd.read_csv("Resources/Crime Information 2011.csv")
crime_2012 = pd.read_csv("Resources/Crime Information 2012.csv")
crime_2013 = pd.read_csv("Resources/Crime Information 2013.csv")
crime_2014 = pd.read_csv("Resources/Crime Information 2014.csv")
crime_2015 = pd.read_csv("Resources/Crime Information 2015.csv")
crime_2016 = pd.read_csv("Resources/Crime Information 2016.csv")
crime_2017 = pd.read_csv("Resources/Crime Information 2017.csv")
crime_2018 = pd.read_csv("Resources/Crime Information 2018.csv")
crime_2019 = pd.read_csv("Resources/Crime Information 2019.csv")
crime_2020 = pd.read_csv("Resources/Crime Information 2020.csv")
```

```python
# Function to create Total Crime column and return DataFrame with just Neighborhood Name and Total Crime
def Total_Crime(year, df):
    df["TOTAL_CRIME"] = (df["HOMICIDE"] + df["SEX_OFFENSE"] + df["ROBBERY"] + 
                             df["ASSAULT"] + df["BURGLARY"] + df["VICE"] + 
                             df["THEFT"] + df["VEH_THEFT"] + df["OTHER"])
    df = df.rename(columns={"TOTAL_CRIME": f"{year}_TOTAL_CRIME"})
    df = df[["NEIGHBORHOOD_NAME", f"{year}_TOTAL_CRIME"]]
    return df
```

```python
# Call Total_Crime function for each year
crime_2000 = Total_Crime("2000", crime_2000)
crime_2001 = Total_Crime("2001", crime_2001)
crime_2002 = Total_Crime("2002", crime_2002)
crime_2003 = Total_Crime("2003", crime_2003)
crime_2004 = Total_Crime("2004", crime_2004)
crime_2005 = Total_Crime("2005", crime_2005)
crime_2006 = Total_Crime("2006", crime_2006)
crime_2007 = Total_Crime("2007", crime_2007)
crime_2008 = Total_Crime("2008", crime_2008)
crime_2009 = Total_Crime("2009", crime_2009)
crime_2010 = Total_Crime("2010", crime_2010)
crime_2011 = Total_Crime("2011", crime_2011)
crime_2012 = Total_Crime("2012", crime_2012)
crime_2013 = Total_Crime("2013", crime_2013)
crime_2014 = Total_Crime("2014", crime_2014)
crime_2015 = Total_Crime("2015", crime_2015)
crime_2016 = Total_Crime("2016", crime_2016)
crime_2017 = Total_Crime("2017", crime_2017)
crime_2018 = Total_Crime("2018", crime_2018)
crime_2019 = Total_Crime("2019", crime_2019)
crime_2020 = Total_Crime("2020", crime_2020)

# Merge the crime data
merged_crime_df = pd.merge(crime_2000, crime_2001, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2002, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2003, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2004, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2005, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2006, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2007, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2008, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2009, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2010, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2011, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2012, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2013, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2014, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2015, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2016, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2017, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2018, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2019, on="NEIGHBORHOOD_NAME")
merged_crime_df = pd.merge(merged_crime_df, crime_2020, on="NEIGHBORHOOD_NAME")

# Add a new column that gives the mean of all the years
merged_crime_df["MEAN_TOTAL_CRIME"] = (
    (merged_crime_df["2000_TOTAL_CRIME"] + merged_crime_df["2001_TOTAL_CRIME"] + merged_crime_df["2002_TOTAL_CRIME"] + 
    merged_crime_df["2003_TOTAL_CRIME"] + merged_crime_df["2004_TOTAL_CRIME"] + merged_crime_df["2005_TOTAL_CRIME"] + 
    merged_crime_df["2006_TOTAL_CRIME"] + merged_crime_df["2007_TOTAL_CRIME"] + merged_crime_df["2008_TOTAL_CRIME"] + 
    merged_crime_df["2009_TOTAL_CRIME"] + merged_crime_df["2010_TOTAL_CRIME"] + merged_crime_df["2011_TOTAL_CRIME"] + 
    merged_crime_df["2012_TOTAL_CRIME"] + merged_crime_df["2013_TOTAL_CRIME"] + merged_crime_df["2014_TOTAL_CRIME"] + 
    merged_crime_df["2015_TOTAL_CRIME"] + merged_crime_df["2016_TOTAL_CRIME"] + merged_crime_df["2017_TOTAL_CRIME"] + 
    merged_crime_df["2018_TOTAL_CRIME"] + merged_crime_df["2019_TOTAL_CRIME"] + merged_crime_df["2020_TOTAL_CRIME"])/21
)

# Show new DataFrame
merged_crime_df
```
