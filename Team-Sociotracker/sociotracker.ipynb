```python
# Function to create a single DataFrame from crime data based on Neighborhood
def crime_data(neighborhood):
    years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 
             2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
    neighborhood_df = pd.DataFrame()
    neighborhood_df = crime_2000.loc[(crime_2000["NEIGHBORHOOD_NAME"] == neighborhood)]
    neighborhood_df = neighborhood_df.append(crime_2001.loc[(crime_2001["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2002.loc[(crime_2002["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2003.loc[(crime_2003["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2004.loc[(crime_2004["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2005.loc[(crime_2005["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2006.loc[(crime_2006["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2007.loc[(crime_2007["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2008.loc[(crime_2008["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2009.loc[(crime_2009["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2010.loc[(crime_2010["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2011.loc[(crime_2011["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2012.loc[(crime_2012["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2013.loc[(crime_2013["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2014.loc[(crime_2014["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2015.loc[(crime_2015["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2016.loc[(crime_2016["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2017.loc[(crime_2017["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2018.loc[(crime_2018["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2019.loc[(crime_2019["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df = neighborhood_df.append(crime_2020.loc[(crime_2020["NEIGHBORHOOD_NAME"] == neighborhood)])
    neighborhood_df["YEAR"] = years
    neighborhood_df = neighborhood_df[["YEAR", "NEIGHBORHOOD_NAME", "HOMICIDE", "SEX_OFFENSE", "ROBBERY",
                                       "ASSAULT", "BURGLARY", "VICE", "THEFT", "VEH_THEFT", "OTHER"]]
    neighborhood_df = neighborhood_df.set_index("YEAR")
    
    # Add column for total crime
    neighborhood_df["TOTAL_CRIME"] = (neighborhood_df["HOMICIDE"] + neighborhood_df["SEX_OFFENSE"] + neighborhood_df["ROBBERY"] + 
                                      neighborhood_df["ASSAULT"] + neighborhood_df["BURGLARY"] + neighborhood_df["VICE"] + 
                                      neighborhood_df["THEFT"] + neighborhood_df["VEH_THEFT"] + neighborhood_df["OTHER"])
    
    # Add column for percent change year to year
    prior_year = 0
    neighborhood_df["CRIME_PERCENT_CHANGE"] = ""
    for index, row in neighborhood_df.iterrows():
        if prior_year == 0:
            neighborhood_df.loc[index, "CRIME_PERCENT_CHANGE"] = 0
            prior_year = row["TOTAL_CRIME"]
        else:
            neighborhood_df.loc[index, "CRIME_PERCENT_CHANGE"] = ((row["TOTAL_CRIME"] - prior_year)/prior_year)*100
            prior_year = row["TOTAL_CRIME"]
    
    # Get rid of trailing zeroes
    neighborhood_df["HOMICIDE"] = neighborhood_df["HOMICIDE"].astype(int)
    neighborhood_df["SEX_OFFENSE"] = neighborhood_df["SEX_OFFENSE"].astype(int)
    neighborhood_df["ROBBERY"] = neighborhood_df["ROBBERY"].astype(int)
    neighborhood_df["ASSAULT"] = neighborhood_df["ASSAULT"].astype(int)
    neighborhood_df["BURGLARY"] = neighborhood_df["BURGLARY"].astype(int)
    neighborhood_df["VICE"] = neighborhood_df["VICE"].astype(int)
    neighborhood_df["THEFT"] = neighborhood_df["THEFT"].astype(int)
    neighborhood_df["VEH_THEFT"] = neighborhood_df["VEH_THEFT"].astype(int)
    neighborhood_df["OTHER"] = neighborhood_df["OTHER"].astype(int)
    neighborhood_df["TOTAL_CRIME"] = neighborhood_df["TOTAL_CRIME"].astype(int)
    return neighborhood_df
```

```python
# Function to transform Zillow data to cover a single Neighborhood from 2000 to 2020
def zillow_data(neighborhood):
    years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 
             2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
    zillow_neighborhood_df = pd.DataFrame()
    zillow_neighborhood_df = zillow_df.loc[(zillow_df["Region Name"] == neighborhood)]
    zillow_neighborhood_df = zillow_neighborhood_df.set_index("Region Name")
    prices = [zillow_neighborhood_df.loc[neighborhood, "2000"],
              zillow_neighborhood_df.loc[neighborhood, "2001"],
              zillow_neighborhood_df.loc[neighborhood, "2002"],
              zillow_neighborhood_df.loc[neighborhood, "2003"],
              zillow_neighborhood_df.loc[neighborhood, "2004"],
              zillow_neighborhood_df.loc[neighborhood, "2005"],
              zillow_neighborhood_df.loc[neighborhood, "2006"],
              zillow_neighborhood_df.loc[neighborhood, "2007"],
              zillow_neighborhood_df.loc[neighborhood, "2008"],
              zillow_neighborhood_df.loc[neighborhood, "2009"],
              zillow_neighborhood_df.loc[neighborhood, "2010"],
              zillow_neighborhood_df.loc[neighborhood, "2011"],
              zillow_neighborhood_df.loc[neighborhood, "2012"],
              zillow_neighborhood_df.loc[neighborhood, "2013"],
              zillow_neighborhood_df.loc[neighborhood, "2014"],
              zillow_neighborhood_df.loc[neighborhood, "2015"],
              zillow_neighborhood_df.loc[neighborhood, "2016"],
              zillow_neighborhood_df.loc[neighborhood, "2017"],
              zillow_neighborhood_df.loc[neighborhood, "2018"],
              zillow_neighborhood_df.loc[neighborhood, "2019"],
              zillow_neighborhood_df.loc[neighborhood, "2020"]]
    zillow_multirow_neighborhood_df = pd.DataFrame()
    zillow_multirow_neighborhood_df["YEAR"] = years
    zillow_multirow_neighborhood_df["NEIGHBORHOOD_NAME"] = neighborhood
    zillow_multirow_neighborhood_df["AVERAGE_HOUSE_PRICE"] = prices
    zillow_multirow_neighborhood_df = zillow_multirow_neighborhood_df.set_index("YEAR")

    # Add column for percent change year to year
    prior_year = 0
    zillow_multirow_neighborhood_df["PRICE_PERCENT_CHANGE"] = ""
    for index, row in zillow_multirow_neighborhood_df.iterrows():
        if prior_year == 0:
            zillow_multirow_neighborhood_df.loc[index, "PRICE_PERCENT_CHANGE"] = 0
            prior_year = row["AVERAGE_HOUSE_PRICE"]
        else:
            zillow_multirow_neighborhood_df.loc[index, "PRICE_PERCENT_CHANGE"] = ((row["AVERAGE_HOUSE_PRICE"] - prior_year)/prior_year)*100
            prior_year = row["AVERAGE_HOUSE_PRICE"]

    zillow_multirow_neighborhood_df["AVERAGE_HOUSE_PRICE"] = zillow_multirow_neighborhood_df["AVERAGE_HOUSE_PRICE"].astype(int)
    return zillow_multirow_neighborhood_df
```
